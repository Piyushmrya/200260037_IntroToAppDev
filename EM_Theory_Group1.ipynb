#**Introduction**

In this passage we discuss the complex Maxwell stress tensor and its importance in understanding electromagnetic forces in optical manipulation. The real part of the Maxwell stress tensor is well-known and describes the time-averaged or real Lorentz forces on a volume of charges and currents. However, the imaginary part of the complex Maxwell stress tensor, related to the exchange of reactive momentum, has so far been ignored. In this report we explain the complex Maxwell stress tensor theorem and understand the concept of the imaginary Lorentz force and the reactive strength of orbital momentum. This reactive force is not observable on time averaging since its net value is zero, but it exists instantaneously due to the transfer of the reactive momentum, which alternates with time, between the wave and the body.
Hence, the ILF is a basically fundamental dynamic phenomenon, inherent to the emergence of  electromagnetic optical forces, being also
associated to the appearance of reactive energy,reactive work, and reactive helicity.  The RLF having been for many years a well-known workhorse in the design of RF antennas, and recently studied in nano and micro-antennas etc. Therefore, like in RF antenna design one aims to diminish the reactive power and reactive work to increase the radiation efficiency, the theory put forward in this work constitutes a tool to act on the ROM and ILF in order to optimize a desired radiation pressure in optical manipulation.

# Complex Maxwell Stress Tensor Theorem , Background and beyond

In the paper, a homogenous medium with $\epsilon = \mu = 1$ in Gaussian units with a body on which studies are undertaken. In the paper, to study the imaginary effects, all the physical quantities are analytically continued in the lower half complex plane $\tau=t-is$.\
We start with analytic signals, $\vec{\mathcal{E}} (r,t) ,\vec{\mathcal{H}}(r,t),\vec{\mathcal{J}}(r,t)$  associated to the real vectors $\vec{E}(r,t) ,\vec{H}(r,t),\vec{J}(r,t)$ which are analytically continued into the lower half complex plane τ = t − is. Generically
denoting each of these analytic functions as $\vec{\mathcal{V}}(r,t,s)$ , 
they are expressed by the Fourier integral : \
\begin{equation} \quad \vec{\mathcal{V}}(r,t,s) \equiv \vec{\mathcal{V}}(r,\tau) =  \int_{0}^{\infty} exp(-i\omega\tau) \vec{\mathcal{D}}(r,\omega) \,d\omega 
\end{equation}
$ \qquad $ where , $ \vec{\mathcal{D}}(r,\omega) $ is  $  \omega $ - Fourier spectrum of the real function $ \vec{\mathcal{D}}(r,t)$ which generally denotes $ \qquad \vec{E}(r,t) ,\vec{H}(r,t) $ or $\vec{J}(r,t)$

Complex derivatives in terms of $\tau$:
\begin{equation}
    \partial_{\tau} = \frac{1}{2}(\partial_t + i\partial_s), \quad
     \partial_{{\tau}}^* = \frac{1}{2}(\partial_t - i\partial_s) \quad.....(1)
\end{equation} 

\begin{equation}
    \partial_{\tau}\mathbf{V^*} = \partial_t \mathbf{V^*} = i\partial_s \mathbf{V^*}, \quad
    \partial_{{\tau}}^*\mathbf{V} = \partial_t \mathbf{V} = - i\partial_s \mathbf{V} \quad.....(2)
\end{equation}

Using eqautions (1) and (2), writing Maxwell Equations for complex analytical signals:

\begin{equation}
    \nabla \cdot \mathbf{E} = 4\pi\rho, \quad
    \nabla \cdot \mathbf{H} = 0, \quad
    \nabla\times\mathbf{E} = -\frac{1}{c}{\partial_\tau^*\mathbf{H}}, \quad
    \nabla\times\mathbf{H} = \frac{1}{c}{\partial_\tau^*\mathbf{E}} + \frac{4\pi}{c}\mathbf{J}\quad .....(3)
\end{equation}

Now, in the paper, it was argued that for complex Lorentz force $\mathbf{F}$ is taken from complex sources. This complex force can be written in terms of conservation of complex linear momentum:
\begin{equation}
    \mathbf{F}(\mathbf{r},t,s) = \frac{\partial}{\partial t}\mathbf{P}_{mech}(\mathbf{r},t,s) = \frac{1}{2}\int_V d^3r'\left[\rho^*(\mathbf{r}',t)\mathbf{E}(\mathbf{r}',t) + \frac{\mathbf{J^*}(\mathbf{r}',t)}{c}\times\mathbf{B}(\mathbf{r}',t)\right] \quad.....(4)
\end{equation}

Substituting $\rho^*$ and $\mathbf{J^*}$ from Maxwell equations (3) and using equations (1), (2) and also, we introduce scaled Poynting momentum $\mathbf{G} = \frac{1}{c^2}\mathbf{S} = \frac{1}{8\pi c}[\mathbf{E} \times \mathbf{B^*}]$. After all this and some mathematics, paper takes it to this:

\begin{equation}
F_i = \partial_t P_{mech,i} = -\int_V d^3r \partial_tG_i^* + \int_{\partial V} d^2r T_{ij}n_j + \frac{i}{8\pi}\int_V d^3r Im[B_j^*\partial_i B_j - E_j^*\partial_iE_j] \quad.....(5)
\end{equation}
where $n_j$ denotes jth Cartesian component of unit outward normal to the surface V, and $T_{ij}$ is Complex Maxwell Stress Tensor given by:
\begin{equation}
T_{ij}(\mathbf{r},s,t) = \frac{1}{8\pi}\left(E_i(\mathbf{r},s,t)E_j^*(\mathbf{r},s,t) + B_i(\mathbf{r},s,t)B_j^*(\mathbf{r},s,t) - \frac{1}{2}\delta_{ij}(|\mathbf{E}|^2 + |\mathbf{B}|^2)\right)\quad.....(6)
\end{equation}

In equation (5), for the force, there appears orbital momentum due to EM field. The orbital momentum of EM field:
\begin{equation}
\mathbf{P^O}=\frac{1}{2}(\mathbf{P^O_e}+\mathbf{P^O_m})\\
(P^O_e)_i = \frac{1}{8\pi \omega}Im[E^*_j\partial_iE_j] \quad (P^O_m)_i = \frac{1}{8\pi \omega}Im[B^*_j\partial_iB_j]\\
\omega=kc \quad .....(7)
\end{equation}
Now, since 3rd term of equation (5) is completely imaginary, the breakdown of force into real and imaginary parts is as follows:

\begin{equation}
F^R_i = -\int_V d^3r \partial_t G^R_i + \int_{\partial V}d^2r T^R_{ij}n_j \quad.....(8)
F^I_i = -\int_V d^3r \partial_t G^R_i + \int_{\partial V}d^2r T^I_{ij}n_j+\omega \int_V d^3r[\mathbf{P^O_m}-\mathbf{P^O_e}]_i \quad.....(9)
\end{equation}
The superscripts R and I denotes real and imaginary parts respectively. And subscript i denotes ith component.

The equation (9) gives us the Imaginary Lorentz Force (ILF) acting on the particle. It comes from time change of IPM, the flux of IMST over the volume in which particle is stored, term containing integral of difference of magnetic and electric orbital momentum, this term, paper has defined to be reactive strength of orbital momentum (ROM) stored in volume V around the particle.

$\mathbf{F^I}$, ILF, also called as reactive force on object of volume $V_O$, has zero time average. On the other hand, RLF is not zero on time averaged scale.


\\

Now, paper talks about specific expressions for EM fields, here **Time harmonic fields**:

So, 
\begin{equation}
\mathbf{E} = \mathbf{E(r)}exp(-i\omega t) \quad \mathbf{B} = \mathbf{B(r)}exp(-i\omega t) 
\end{equation}

In this case the time derivate of complex poynting momentum $[\mathbf{E \times B^*}]$ term will go to zero and hence, the force equation (5) will here become:

\begin{equation}
F_i = \int_{\partial V} d^2r T_{ij}n_j + i\omega\int_V d^3r (\mathbf{P^O_m-P^O_e)_i}) .....(10)
\end{equation}

For, the real part of force, we take time-average, where second term will go to zero, and hence:

\begin{equation}
Re\{F_i\}=<F_i> = \int_{\partial V} d^2r <T_{ij}>n_j \quad.....(11)
\end{equation}
where $<T_{ij}>$ is just the real part of $T_{ij}$.

The imaginary part of force equation (10) will be:

\begin{equation}
Im\{F_i\}=F^I_i = \int_{\partial V} d^2r T_{ij}^In_j +\omega\int_V d^3r (\mathbf{P^O_m-P^O_e)_i}\quad.....(12)
\end{equation}

**The canonical and spin momenta and the complex lorentz force** \\
In this section we show the relation between complex lorentz force and the momentas. Using following identities : 
\begin{equation}
a^{*} \times ( Δ \times a ) = a^{*}_{j} \partial_{i} a_{j} - a^{*}_{j} \partial_{j} a_{i} 
\end{equation}
and Levi-Civita Tensor:
\begin{equation}
ϵ_{ijk} ϵ_{klm} = δ_{il}δ_{jm} - δ_{im}δ_{jl}
\end{equation}
Hence, using the above identity, we may write these spin
momentum densities as : 
\begin{equation}
   P^{S}_{e} = \frac{1}{16Πω}\ 𝚫×IM({E^{*}×E}) = \frac{1}{16Πω}\ 𝚫×I(\partial_{j} E^{*}_{i}E_{j}) 
\end{equation}
\begin{equation}
   P^{S}_{m} = \frac{1}{16Πω}\ 𝚫×IM({B^{*}×B}) = \frac{1}{16Πω}\ 𝚫×I(\partial_{j} B^{*}_{i}B_{j}) \quad .....(26) 
\end{equation}    


Therefore from eqn (21) we finally have 
\begin{equation}
  \mathcal{F}^{I} = ω \int_{V}  d^{3}r [(P^{S}_{m} - P^{S}_{e}) + (P^{O}_{m} - P^{O}_{e})] \quad .....(28)

# Example 2: Linearly polarized plane wave impinging on a low refractive index small dielectric particle

Here, the paper has an example of a purely dielectric particle sphere, with following parameters:\
Refractive index **n = 1.59** \
Radius **a = 50 nm**

The particle is illuminated by a linearly propagating plane wave:
\
\
\begin{align}
\textbf{E} = E_0e^{ikz} \hat{x}, \quad \textbf{B} = E_0e^{ikz} \hat{y} \quad .....(1)
\end{align}
\
From supplementary material of the paper, in appendix E, the time-averaged and the reactive force on the particle are computed for a dipole, in this case electric dipole:
\
\begin{equation}
<F_i>=\frac{1}{2}Re\{\alpha_eE_j\partial_iE_j^*\}, \quad Im\{F_i\}=-\frac{1}{2}Im\{\alpha_eE_j\partial_iE_j^*\}
\end{equation}
\
Using given EM fields from equation (1), real force (time-averaged) and imaginary force are:
\begin{align}
<\mathcal{\textbf{F}}>=\frac{1}{2}\alpha_e^IkE_0^2 \hat{z}, \quad \textbf{F}_e^I=\frac{1}{2}\alpha_e^RkE_0^2\hat{z} \quad .....(1)
\end{align}
\
Now, how to find real and imaginary polarizibilities, $\alpha_e^R$ and $\alpha_e^I$ respectively.

We use Clausius-Mossotti relation between electric polarisibility and refractive index:
\
\
\begin{align}
\alpha_e = \frac{3}{4\pi}{V}(\frac{n^2-1}{n^2+2})
\end{align}
\
Where $V$ is volume of the particlem which is $\frac{4\pi}{3}a^3$. Now, since $n$ has no imaginary part, $\alpha_e^I$ will be zero. So, we will get $\alpha_e^R$ from the equation.
\
Plotting $k\alpha_e^R$ vs wavelength $\lambda$, where $k=\frac{2\pi}{\lambda}$ in figure 1.

Now, for ILF, using equation 1, taking $E_0 = 1$:  \\
$\mathbf{F^I_e} = \frac{1}{2}k\alpha_e^R \hat z$ and since $\alpha^I_e=0 \implies <\mathbf{F}>=0$  
Plotting imaginary and real forces for this particle in figure 2.

In the paper, imaginary lorentz force (ILF) is broken into $F^{IMST}$ and $F^{ROM}$. For the former, we need to solve equation the first term in RHS of equation (12) in Complex Maxwell Stress Theorem section, which is as follows:
\begin{equation}
Im\{F_i\}=F^I_i = \int_{\partial V} d^2r T_{ij}^In_j +\omega\int_V d^3r (\mathbf{P^O_m-P^O_e)_i}\quad \\
T_{ij}^I = \frac{1}{8\pi} Im\{E_i E_j^* + B_j B_i^*\}
\end{equation}

Solving this is not trivial, since electric fields in the vicinity of particle will change, so appendix D in supplementary material has given an approximated solution for $F^{IMST}$, which is given in equation (55) in paper:
\begin{equation}
F^{IMST}=Im\left\{\int_{\partial V_0} d^2\mathbf{r}\, T^{\mathrm{mix}}_{kj}n_j \right\}= -\frac{1}{10}\{\alpha_{e}^I|\nabla\mathbf{E}|^2 - \alpha_{m}^I|\nabla\mathbf{B}|^2\} + \frac{4\pi}{5}\omega\{(\alpha^R_m - \alpha^R_e)\langle \mathbf{g}\rangle + (\alpha_{e}^I-\alpha_{m}^I)\mathbf{g^I} + 2(\alpha_e^R P_e^O - \alpha_m^R P_m^O)\} \\
<g> = \frac{1}{8\pi c}Re\{\mathbf{E\times B^*}\}
\end{equation}


Since, particle in this example is purely dielectric, $\alpha_m = 0$, using this and other values from the example, it comes to be:

\begin{equation}
F^{IMST} = \frac{1}{5} \{ \frac{1}{2} \alpha^R_e k |E_0|^2 \} = \frac{1}{5} F_e^I
\end{equation}

So, it comes to be one fifth of ILF.

Now, calculating force due to reactive orbital momentum storage is easy, we just subtract $F^{IMST}$ from ILF.

Both of them are plotted for this example in figure 1(c) and 1(d) respectively.

```python
import matplotlib.pyplot as plt
import numpy as np
n = 1.59
a = 50
wavelength = np.arange(450,1100)
k_alp = (a**3)*((n**2-1)/(n**2+2))*(2*3.141/wavelength)
F_im = (1/2)*k_alp
F_imst = (1/5)*F_im
F_rom = F_im - F_imst
figure, axis = plt.subplots(2, 2)
axis[0,0].plot(wavelength,k_alp)
axis[0,0].set_title(r'$1(a)\quad k\alpha_e^R$ vs $\lambda (nm)$')
axis[1,0].plot(wavelength,F_im, color='r')
axis[1,0].set_title(r'1(b) $\quad F^I_e$ vs $\lambda$')
axis[0,1].plot(wavelength,F_imst)
axis[0,1].set_title(r'$1(c)\quad F^{IMST}$ vs $\lambda (nm)$')
axis[1,1].plot(wavelength,F_rom)
axis[1,1].set_title(r'$1(d)\quad F^{ROM}$ vs $\lambda (nm)$')
figure.tight_layout(pad=1.0)
plt.show()
```

# Tutorial Question : Linearly polarized Gaussian wave impinging on a low refractive index small dielectric particle

##Solution

Given \begin{equation}
  \mathbf{E}=E_0 exp\left(-\frac{R^2}{\sigma^2}\right)exp(ikz) \hat x, \quad (R=\sqrt{x^2+y^2})\quad .....(1)
\end{equation} 
And similar form of magnetic field.\
In general, Force on moving charge particle is given by Lorentz force 
\begin{equation}
 \mathcal{F} = q ( \vec{E} + \vec{v}\times \vec{B}) 
\end{equation}
In our problem we have linearly polarized gaussian beam incident on small particle which is **electrically dipolar**.\
A dipole is the limit of two charges of equal magnitude and opposite sign spaced a distance d apart, in the limit 
\begin{equation}
\lim_{d \to 0} (qd)  =  p
\end{equation}
so the force experienced is 
\begin{equation}
\mathcal{F}_{e} = q (\vec{E}(r+d) - \vec{E}(r))=(\vec{p} · Δ)\vec{E} \\
\mathcal{F}_{m} = q(\vec{v}\times \vec{B}) = \frac{1}{c} \frac{d\vec{p}}{dt} × \mathcal{\vec{B}}
\end{equation}
So Finally we have 
\begin{equation}
\mathcal{F} = ( \vec{p} · Δ) \vec{E} + \frac{1}{c} \frac{d\vec{p}}{dt} × \mathcal{\vec{B}}
\end{equation}
Putting $\vec{E} $ and $ \vec{B}$
\begin{equation}
F = ( \vec{p} · Δ) E + i k \frac{d\vec{p}}{dt} × \mathcal{B}=\frac{1} {2}[α_{e} E{j}^{*}∂_{j}E_{i} - ik α^{*}_{e} ϵ_{ijk} E_{j}^{*} B_{k}]
\end{equation}

This example is similar to example 2 except it has a gaussian profile. 

Using the equation for force on electric dipole from appendix E, 
\begin{equation}
<F_i>=\frac{1}{2}Re\{\alpha_eE_j\partial_iE_j^*\}, \quad Im\{F_i\}=-\frac{1}{2}Im\{\alpha_eE_j\partial_iE_j^*\}
\end{equation}

Solving for x-component of time-averaged force:
\begin{equation}
<F_x>=\frac{1}{2} Re\{\alpha_e(E_x\partial_xE_x^*+E_y\partial_xE_y^*+E_z\partial_xE_z^*)\}
\\
E_y=E_z=0
\\
\partial_x E_x^*=\partial_x\left\{E_0^* exp\left(-\frac{x^2+y^2}{\sigma^2}\right)exp(-ikz)\right\} \\
\implies \partial_x E_x^*=E_0^*\left(-\frac{2x}{\sigma^2}\right)exp\left(-\frac{x^2+y^2}{\sigma^2}\right)exp(-ikz) \\
\implies E_x\partial_x E_x^*=|E_0|^2\left(-\frac{2x}{\sigma^2}\right)exp\left(-\frac{2x^2+2y^2}{\sigma^2}\right) \\
\implies <F_x>=\frac{1}{2} Re\{\alpha_e(E_x\partial_xE_x^*)\}=\frac{1}{2}\alpha_e^R |E_0|^2\left(-\frac{2x}{\sigma^2}\right)exp\left(-\frac{2x^2+2y^2}{\sigma^2}\right)\\
\implies <F_x>=-\alpha_e^R |E_0|^2\left(\frac{x}{\sigma^2}\right)exp\left(-\frac{2R^2}{\sigma^2}\right)\\
\end{equation}
Similarly, for y-component of force:
\begin{equation}
<F_y>=-\alpha_e^R |E_0|^2\left(\frac{y}{\sigma^2}\right)exp\left(-\frac{2R^2}{\sigma^2}\right)\\
\end{equation}
Now, for z-component:
\begin{equation}
<F_z>=\frac{1}{2} Re\{\alpha_e(E_x\partial_zE_x^*)\}\\
\implies \partial_z E_x^*=E_0^*\left(-ik\right)exp\left(-\frac{R^2}{\sigma^2}\right)exp(-ikz) \\
\implies E_x\partial_z E_x^*=|E_0|^2\left(-ik\right)exp\left(-\frac{2R^2}{\sigma^2}\right) \\
\implies
<F_z>=\frac{1}{2} Re\{\alpha_e(E_x\partial_zE_x^*)\}=\frac{1}{2}k\alpha^I_e|E_0|^2exp\left(-\frac{2R^2}{\sigma^2}\right)\\
\end{equation}
This gives real force:
\begin{equation}
<\mathbf{F}>=|E_0|^2exp\left(-\frac{2R^2}{\sigma^2}\right)\left[-\frac{\alpha^R_e}{\sigma^2}(\mathbf{x}+\mathbf{y})+\frac{k\alpha^I_e}{2}\mathbf{\hat z}\right]
\end{equation}
Now, solving for imaginary part of force:
\begin{equation}
Im\{\mathbf{F}\} = |E_0|^2exp\left(-\frac{2R^2}{\sigma^2}\right)\left[\frac{\alpha^I_e}{\sigma^2}(\mathbf{x}+\mathbf{y})+\frac{k\alpha^R_e}{2}\mathbf{\hat{z}}\right]
\end{equation}

# The time-averaged Poynting momentum with sources and the imaginary Lorentz force

 We previosly saw that the imaginary lorentz force is related 
 \begin{equation}
  \mathbf{F}^I=ω\int_Vd^3r[(P^S_m-P^S_e)+(P^o_m-P^o_e)]
 \end{equation}
 Let us now establish the compatibility of time averaged poynting momentum with the imaginary lorentz force.We know the definition of time averaged poynting momentum.
 \begin{equation}
 <g>=\frac{<S>}{c^2}=(\frac{1}{8πc})Re{(\mathbf{E}×\mathbf{B^*})}
 \end{equation}

 Using the langrangian formulation in electrodynamics the above equation can be written completely in terms of electric field as follows:
 \begin{equation}
 <g>=P^S_m+P^O_e+\frac{1}{2ω}Im({ρ^*\mathbf{E}})\quad.....1
 \end{equation}
 Similarly eliminating electric field in the definition of $<g>$, we obtain in terms of magnetic momenta we get
 \begin{equation}
 <g>=P^S_m+P^O_m-\frac{1}{2ω}Im(\mathbf{j^*}×\mathbf{B})\quad......2
 \end{equation}
 Adding the above two equations and considering there is no source we get
 \begin{equation}
 <g>=(P^O_e+P^O_m)/2+(P^S_e+P^S_m)/2
 \end{equation}
 Subtracting equation 1 from 2 and integrating in V we get the imaginary lorentz force in terms in spin and canonical momentum.
 Similarly if we do this for the imaginary field momentum following the same procedure we get following two important equations
 \begin{equation}
 <\mathbf{F}>==2ω\int_V d^3r(P^{SI}+P^{OI})
 \end{equation}
 where $P^{SI}=\frac{P^{SI}_e+P^{SI}_m}{2}$ and $P^{OI}=\frac{P^{OI}_e+P^{OI}_m}{2}$
 Above equation is the equation for Real lorentz force in terms of imaginary mowmnta.
 The second equation is for imaginary poynting momentum in terms of reactive spin and orbital momentum
 \begin{equation}
 \mathbf{g}^I=\mathbf{P^S+P^O}+\frac{1}{4kc}Re({\mathbf{ρ^*}\mathbf{E}-\frac{1}{c}\mathbf{j^*}\times \mathbf{B}}
 \end{equation})
 where $\mathbf{P}^S=\frac{\mathbf{P}^{SI}_m-\mathbf{P}^{SI}_e}{2}$ and $\mathbf{P}^O=\frac{\mathbf{P}^{OI}_m-\mathbf{P}^{OI}_e}{2}$
